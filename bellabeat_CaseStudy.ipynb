### Visualization of analysis

<h6 style="font-size:2rem;color:#888888;">dailyActivity</h6>

 ###### Average Total Steps vs. Average Calories burned per day

```python
from plotnine import ggplot, aes, geom_point, labs, theme_minimal, scale_color_gradient, theme, element_text

# Load data from CSV file
data = pd.read_csv('bellabeat_data - dailyActivity.csv')

# Create the scatter plot using plotnine
p = (ggplot(data, aes(x='TotalSteps', y='Calories', color='Id')) +
     geom_point() +
     scale_color_gradient(low='pink', high='red') +
     labs(title='Total Steps vs. Calories Burned',
          subtitle='Relationship between calories burned per total step',
          x='Total Steps',
          y='Calories') +
     theme_minimal() +
     theme(plot_title=element_text(hjust=0.5),
           plot_subtitle=element_text(hjust=0.5)))  # Center both title and subtitle

# Display the plot
print(p)
```

From my analysis, a noticeable correlation exists between the total daily steps and the total daily calories burned. The data illustrates that participants who took more steps per day also exhibited higher calorie expenditure.

<h6 style="font-size:2rem;color:#888888;">minuteSleep</h6>

###### Average Restful Moments Throughout the Day

```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, labs, scale_fill_gradient, theme_minimal, theme, element_text

# Load data from CSV file
data = pd.read_csv('bellabeat_sleep - minuteSleepV2.csv')

# Filter data to include only distinct values from the 'Id' column
distinct_data = data.drop_duplicates(subset='Id')

# Create the bar chart using plotnine
p = (ggplot(distinct_data, aes(x='TimeOfDay', y='Value', fill='Id')) +
     geom_bar(stat='identity') +
     scale_fill_gradient(low='lightblue', high='darkblue') +  # Adjust color scale
     labs(title='Times of Rest',
          x='Time of Day',
          y='Value') +
     theme_minimal() +
     theme(axis_text_x=element_text(angle=45, hjust=1),  # Rotate x-axis labels
           plot_title=element_text(hjust=0.5),  # Center the title
           legend_position='right'))  # Move the legend to the right

# Display the plot
print(p)
```

Referring to the Fitbit blog's sleep classifications (Value 1 as non-restful, Value 2 as light sleep, and Value 3 as deep sleep, the latter two indicating better sleep), the data highlights a trend: participants generally experience improved sleep from 8pm to 12:30am. Within this period, participants tend to have periods of wakefulness, indicating some nighttime awakenings on average.
